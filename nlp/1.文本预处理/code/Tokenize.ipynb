```python
def tokenize(content, token='word'):
    """将文本内容拆分成单词或字符词元"""
    if token=='word':
        return content.split()
    elif token == 'char':
        return list(content)
    else:
        print("错误，未知词元类型" + token)
    
    
tokenize(content, token='char')
```
