# 构建词表(词典/字典)Vocab

## 字典中每个词(word) 或 字符(char),都是独一无二的(unique)，每一个token对应有一个index下标，且一般情况下按照出现频率排序，频率高的在前，频率低的在后

### 统计词元频率

```python
from collections import Counter
```

```python
def count_corpus(tokens):
    if isinstance(tokens[0], list):
        # 将词元展平成一个列表
        tokens = [token for line in tokens for token in line]
        
    return Counter(tokens)


# tokens_one_list = ['the', 'time', 'machine', 'by', 'h', 'g', 'wells', 'the']    

counter = count_corpus(tokens)
# Counter(tokens_one_list)
counter
```

## 构建字典：1.根据词频进行排序

```python
# counter
counter.items() # 返回可遍历的元组数组

token_freqs = sorted(counter.items(), key=lambda x: x[1], reverse=True)
token_freqs
```

## 2.人为添加的特定字符

```python
reserved_tokens = ['<unk>']
```

## 3.从高频到低频分别赋予index

```python
unique_tokens = reserved_tokens + [token for token, freq in token_freqs]
unique_tokens
```

## 4.获取字典长度

```python
len(unique_tokens)
```
